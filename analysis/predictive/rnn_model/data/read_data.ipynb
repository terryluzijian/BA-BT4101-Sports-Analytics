```python
import numpy as np
import pandas as pd
import os
path = os.path.abspath(os.path.join(os.getcwd(), 'rnn_model\data'))
```

```python
def get_matrix_combination(df, drop_col_name='run_date', groupby_col_name='horse_id'):
    if drop_col_name is not None:
        df_dropped = df.reset_index().drop(drop_col_name, axis=1)
    df_matrix = df_dropped.groupby(groupby_col_name).apply(lambda x: x.as_matrix())
    df_matrix = df_matrix.apply(lambda arr_lst: np.array(list(map(lambda arr: arr[1:], arr_lst))))
    return df_matrix

df = pd.read_csv('race_record_first_included_validation.csv', index_col=0)
df.sort_values(['horse_id', 'run_date'], inplace=True)
df.set_index(['horse_id', 'run_date'], inplace=True)
df = df.astype(float)

target = 'y_run_time_1000'

feature_names = list(filter(lambda col_name: 'y_run' in col_name, df.columns))
feature_names.remove(target)
if target != 'y_run_time_1000':
    feature_names.append('last_run_time')
df.drop(feature_names, axis=1, inplace=True)

df_x = df.drop(target, axis=1)
df_y = df[target]
```

```python
a = get_matrix_combination(df_x)
```
